```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
%matplotlib inline
plt.rcParams["figure.figsize"] = (15,10)
```

```python
def f(x, y):
    return np.sin(2*np.pi*x)*np.sin(4*np.pi*y)-2*np.sin(4*np.pi*x)*np.sin(2*np.pi*y)

x = np.linspace(0, 1, 30)
y = np.linspace(0, .5, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax.view_init(40, 55)
ax.set_title('Plot of A(2,2) - 2A(4,1)')
```

```python
def f1(x, y):
    return (np.sin(2*np.pi*x)*np.sin(4*np.pi*y)-2*np.sin(4*np.pi*x)*np.sin(2*np.pi*y) -
           (np.sin(2*np.pi*x)*np.sin(4*np.pi*y)-2*np.sin(4*np.pi*x)*np.sin(2*np.pi*y))*np.e**(2j))  

x = np.linspace(-0.001, 1.001, 30)
y = np.linspace(-0.001, .5001, 30)

X, Y = np.meshgrid(x, y)
Z1 = np.real(f1(X, Y))
```

```python
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z1, 1)
ax.view_init(88, 90)
ax.set_title('Plot of A(2,2) - 2A(4,1) Nodal Lines')
```
